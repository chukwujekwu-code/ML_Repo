## Note

- We need to drop high cardinality columns
- The price column is the target column

```python
df.hist(bins = 50, figsize= (20, 8))
```

```python
df.plot(kind= "scatter", x = "longitude", y = "latitude")
```

```python
# visualize the geographical data
df.plot(kind= "scatter", x = "longitude", y = "latitude", alpha = 0.1)
```

```python
plt.figure(figsize= (20, 8))
df.plot(kind= "scatter", x = "longitude", y = "latitude", alpha = 0.05,
        c = "price",
        cmap=plt.get_cmap("coolwarm"), colorbar= True)

plt.show()
```

```python
df.dtypes
```

```python
# drop high cardinality columns
cols = ["brokertitle", "address", "main_address", "state", "formatted_address", "long_name", "street_name"]
df.drop(columns= cols, inplace= True)
```
