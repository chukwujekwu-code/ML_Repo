### Cross Validation From Scratch

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone
```

```python
skfolds = StratifiedKFold(n_splits= 3, random_state= 42, shuffle= True)

for train_index, test_index in skfolds.split(X_train, y_train_5):
    clone_clf = clone(sgd_clf)
    X_train_folds = X_train[train_index]
    y_train_folds = y_train_5[train_index]
    X_test_folds = X_train[test_index]
    y_test_folds = y_train_5[test_index]

    clone_clf.fit(X_train_folds, y_train_folds)
    y_pred = clone_clf.predict(X_test_folds)
    n_correct = sum(y_pred == y_test_folds)
    print(n_correct / len(y_pred))
```

How do we decide which threshold to use? 
- Firstly, we use the cross_val_predict() function to get the scores of all instances in the training set.

```python
y_scores = cross_val_predict(
    sgd_clf, X_train, y_train_5,
    cv=3, method= "decision_function"
)
```

```python
from sklearn.metrics import precision_recall_curve
```

```python
# We use pprecision_recall_curve function to compute precision and recall for al possible thresholds
precision_recall_curve(y_train_5, y_scores)
```

```python
precisions, recalls, thresholds = precision_recall_curve(y_train_5, y_scores)
```

```python
# Lets use matplotlib to plot precision and recall functions of the threshold value

def plot_precision_recall_vs_threshold(precisions, recalls, thresholds):
    plt.figure(figsize=[15,6])
    plt.plot(thresholds, precisions[:-1], "b--", label="Precison")
    plt.plot(thresholds, recalls[:-1], "g-", label = "Recall")
    plt.legend()
    plt.grid()
    plt.xlabel("Threshold")
```

```python
plot_precision_recall_vs_threshold(precisions, recalls, thresholds)
```

```python
plt.plot(precisions, recalls)
plt.grid()
plt.title("Precision vs Recall Plot")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.show()
```

Suppose we decide to aim for 90% Precision. One could look up the plot and find that we need to use a threshold of about 8000.
We could be more precise by searching for the lowest threshold that gives one at least 90% precision. using np.argmax()

```python
threshold_90_precision = thresholds[np.argmax(precisions >= 0.90)]
threshold_90_precision
```

```python
thresholds[np.argmax(recalls >= 0.90)]
```

```python
np.argmax(recalls >= 0.90)
```

```python
np.argmax(precisions >= 0.90)
```

```python
y_train_pred_90 = (y_scores >= threshold_90_precision)

# lets check the predictions precision and recall
precision_score(y_train_5, y_train_pred_90)
```

## The ROC Curve

```python
from sklearn.metrics import roc_curve
```

```python
fpr, tpr, thresholds = roc_curve(y_train_5, y_scores)
```

```python
# lets plot FPR against TPR
def plot_roc_curve(fpr, tpr, label = None):
    plt.plot(fpr, tpr, linewidth = 2, label = label)
    plt.plot([0, 1], [0, 1], "k--")
    plt.grid()
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate (Recall)")
    plt.title("The ROC Curve plot")
    plt.tight_layout()
```

```python
plot_roc_curve(fpr, tpr)
plt.show()
```

```python
def plot_digits(instances, images_per_row=10, **options):
    size = 28
    images_per_row = min(len(instances), images_per_row)
    n_rows = (len(instances) - 1) // images_per_row + 1

    # Append empty images to fill the end of the grind, if needed
    n_empty = n_rows * images_per_row - len(instances)
    padded_instances = np.concatenate([instances, np.zeros((n_empty, size * size))], axis=0)

    #Reshape the array so its organized as a grid containing 28x28 images.
    image_grid = padded_instances.reshape((n_rows, images_per_row, size, size))
    big_image = image_grid.transpose(0, 2, 1, 3).reshape(
        n_rows * size, images_per_row * size
    )

    plt.imshow(big_image, cmap= mpl.cm.binary, **options)
    plt.axis("off")
```

```python
cl_a, cl_b = 3, 5 
X_aa = X_train[(y_train == cl_a) & (y_train_pred == cl_a)] 
X_ab = X_train[(y_train == cl_a) & (y_train_pred == cl_b)] 
X_ba = X_train[(y_train == cl_b) & (y_train_pred == cl_a)] 
X_bb = X_train[(y_train == cl_b) & (y_train_pred == cl_b)] 
plt.figure(figsize=(8,8)) 
plt.subplot(221); plot_digits(X_aa[:25], images_per_row=5) 
plt.subplot(222); plot_digits(X_ab[:25], images_per_row=5) 
plt.subplot(223); plot_digits(X_ba[:25], images_per_row=5) 
plt.subplot(224); plot_digits(X_bb[:25], images_per_row=5) 
plt.show()
```

```python
knn_clf = KNeighborsClassifier()
```

```python
params = {
    "n_neighbors" : np.arange(1,11,4),
    "weights": ["uniform", "distance"],
    "algorithm": ["auto"],
    "leaf_size": np.arange(10,50,10),
    }
```

```python
random_search = RandomizedSearchCV(
    knn_clf,
    param_distributions= params,cv=3, n_iter= 10,random_state= 42,
    return_train_score= True,
    n_jobs = 2
)
```
